Imported Libraries

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

Created a neural network of one neuron/layer and one input with keras

```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

Compiled the model with loss as mean square error and optimizer as sgd (a type of gradient descent)

```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

created the sample inputs

```
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

fitted the model with inputs with epochs as 600 (basically number of repetitions each with decreasing loss/error and increased accuracy)

```
model.fit(xs, ys, epochs=600)
```

tested the model by predicting some values

```
print(model.predict([10.0]))
```

```
print(model.predict([5.0]))
```
